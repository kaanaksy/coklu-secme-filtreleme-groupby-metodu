```python
# Sütunları dilimleme işlemi: mesela burada title ile genre arasındaki tüm sütunlar yazdırıldı.

film.loc[0:5, "title":"genre"]
```

```python
# loc metodu ile "sadece" sütunları parçalayarak seçme işlemi: BÜTÜN satırları alarak yukarıdaki işlemin nasıl yapılacağını gördük

film.loc[:, "title":"genre"]
```

```python
#loc metodu ile filtreleme: Genre'si sadece Crime olan verileri sıraladık

film.loc[film.genre == "Crime",]
```

```python
# loc metodu ile çoklu filtreleme: Genre'si Crime olan verilerin sadece title'larını yazdırdık

film.loc[film.genre == "Crime", "title"]
```

```python
# Sütunlarda filtrelemeyi belirli aralıklarda yapma işlemi: title'dan duration'a kadar olan kısmı sıraladık.

film.loc[film.genre == "Crime", "title" : "duration"]
```

```python
# yukarıdaki işlemi birden fazla sütun için yapmak istersek ikinci köşeli parantezi kullanmamız gerekir: Crime'a sahip verileri sadace title
#ve duration sütunlarıyla filtreledik

film.loc[film.genre == "Crime", ["title","duration"]]
```

```python
film.iloc[0:4, :] # 0dan 4e kadar olan bütün satırları alıp, kısıtlama yapmadığımız için bütün sütunları gösterdi.
```

```python
# Bazı kısıtlamalarla filtreleme yapma

(film.duration >= 140) & (film.genre == "Crime")
```

```python
# Boolean değerleri anlamlandırarak yukarıdaki işlemi yapma: TÜM İFADEYİ köşeli paranteze alıp başına veri adımızı yazarak getirdik

film[(film.duration >= 140) & (film.genre == "Crime") | (film.genre == "Action")]
```

```python
# Yukarıdaki işlemi VEYA operatörü ile yapma: Duration'ı 140a eşit büyük olan veya genresi crime olan tüm veriler geldi.

film[(film.duration >= 140) | (film.genre == "Crime")]
```

```python
# farklı türlere ait filtreleme
# isin metodu ile genre sütunu içinte crime drama ve action olan verileri gösterdik

film[film.genre.isin(["Crime", "Drama", "Action"])]
```
